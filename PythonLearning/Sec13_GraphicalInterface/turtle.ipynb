```python
# 从程序代码可以看出，海龟绘图就是指挥海龟前进、转向，海龟移动的轨迹就是绘制的线条。
# 要绘制一个长方形，只需要让海龟前进、右转90度，反复4次。
# 调用width()函数可以设置笔刷宽度，调用pencolor()函数可以设置颜色。
# 更多操作请参考turtle库的说明。
# 绘图完成后，记得调用done()函数，让窗口进入消息循环，等待被关闭。
# 否则，由于Python进程会立刻结束，将导致窗口被立刻关闭。
# turtle包本身只是一个绘图库，但是配合Python代码，就可以绘制各种复杂的图形。
# 例如，通过循环绘制5个五角星：

# coding: utf-8
import turtle
from turtle import *

def drawStar(x, y):
    print(f"绘制五角星 at ({x}, {y})")  # 调试信息
    pu()  # 抬笔
    goto(x, y)  # 移动到起始点
    pd()  # 落笔
    seth(0)  # 设置朝向0度（向右）
    for i in range(5):
        fd(40)  # 前进40像素
        rt(144)  # 右转144度（五角星角度）

# 设置画笔属性
width(2)  # 笔刷宽度
pencolor("red")  # 颜色设为红色

# 绘制5个五角星
for x in range(0, 250, 50):
    drawStar(x, 0)

# 保持窗口打开
print("绘制完成！窗口保持打开...")
done()
```

```python
# coding: utf-8
import turtle as t  # 避免命名冲突，推荐别名

# 设置色彩模式为RGB
t.colormode(255)  # 正确设置RGB模式（0-255）

t.lt(90)  # 左转90度，树向上

lv = 14  # 递归层级
l = 120  # 初始线长
s = 45   # 转角角度

t.width(lv)  # 设置初始笔刷宽度

# 初始化RGB颜色
r = 0
g = 0
b = 0
t.pencolor(r, g, b)  # 设置初始颜色

t.penup()  # 抬笔
t.bk(l)    # 后退
t.pendown()  # 落笔
t.fd(l)    # 前进，画主干

def draw_tree(l, level):
    global r, g, b
    # 保存当前笔刷宽度
    w = t.width()

    # 缩小笔刷宽度
    t.width(w * 3.0 / 4.0)
    # 更新颜色
    r = r + 1
    g = g + 2
    b = b + 3
    t.pencolor(r % 200, g % 200, b % 200)  # 限制RGB在0-199

    l = 3.0 / 4.0 * l  # 缩短线长

    t.lt(s)  # 左转
    t.fd(l)  # 前进画分支

    if level < lv:
        draw_tree(l, level + 1)  # 递归画左分支
    t.bk(l)  # 退回
    t.rt(2 * s)  # 右转2倍角度
    t.fd(l)  # 前进画右分支

    if level < lv:
        draw_tree(l, level + 1)  # 递归画右分支
    t.bk(l)  # 退回
    t.lt(s)  # 恢复角度

    # 恢复笔刷宽度
    t.width(w)

t.speed("fastest")  # 最快速度

print("开始绘制递归树...")
draw_tree(l, 4)  # 开始递归，初始层级4

print("绘制完成！窗口保持打开...")
t.done()  # 保持窗口
```
